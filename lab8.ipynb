```python
#To visualize each statistics one by one kindly run each statement one by one 

import pandas as pd
import numpy as np
df=pd.read_csv('Employees_Absenteeism.csv',sep=',')
#total no of rows and columns
df.shape
#Basic information
df.info()
#Describe the data
df.describe()
#Find the duplicates
df.duplicated().sum()
#unique values
df['EmployeeNumber'].unique()
df['Surname'].unique()
df['GivenName'].unique()
df['Gender'].unique()
df['City'].unique()
df['JobTitle'].unique()
#Find null values
df.isnull().sum()
#Replace null values
df.replace(np.nan,'0',inplace = True)
#Check the changes now
df.isnull().sum()
#Datatypes
df.dtypes
#Correlation 
df.corr()
#skewness
# skewness along the index axis
df.skew(axis = 0, skipna = True)
# find skewness in each row
df.skew(axis = 1, skipna = True)
#Sum of values
df.sum()
#Number of non-NA values
df.count()
# Describe index
df.index
#Describe DataFrame columns
df.columns
#Sort by the values along an axis
df.sort_values(by='GivenName')
```
